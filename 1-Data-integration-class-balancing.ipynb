```python
# CREST crest_lite_1.csv
df1 = pd.read_csv('https://query.data.world/s/5iz7cqrdwstmkmrr2amlih5twvsb5m')
```

```python
#CREST crest_lite_4.csv
df2 = pd.read_csv('https://query.data.world/s/4lmzkr5rrah6icpkfw4nzfigynuvsh')
```

```python
# Keeping relevant features for further exploration

df = df[[ 'title', 'collection', 'document_number', 'release_decision', 'document_page_count',
       'sequence_number', 'publication_date' , 'content_type', 'more1_title', 'more2_title', 'more3_title',
        'more4_title', 'more5_title']]
df.shape
```

```python
# Check the null values
df.isnull().sum().tolist()
```

```python
# Drop null values
df = df.dropna(how="any")
```

```python
# Check the null values
df.isnull().sum().tolist()
```

```python
# check for the class imbalance
df['collection'].value_counts()
```

```python
# merge columns with less values than 10,000 and name it Misc
df['collection'] = df['collection'].map({'General CIA Records': 'General_CIA_Records' ,
                                       'NGA Records (Formerly NIMA)': 'NGA_Records' ,
                                       'Scientific Abstracts' : 'Scientific_Abstracts',
                                       'Consolidated Translations': 'Consolidated_Translations',
                                       'Ground Photo Caption Cards': 'Misc',
                                       'STARGATE' : 'Misc',
                                       'Library of Congress': 'Misc',
                                       'OSS Collection' : 'Misc',
                                       'JPRS' : 'Misc'})
```

```python
# check for the class imbalance after merging
df['collection'].value_counts()
```

```python
# This is a class imbalance problem, let's do undersampling to keep all the values equal for every class

    # use resample method from scikit-learn
    from sklearn.utils import resample

    # Separate majority and minority classes for undersampling
    df_collection_General_CIA_Records              = df[df.collection== 'General_CIA_Records']
df_collection_NGA_Records                      = df[df.collection == 'NGA_Records']
df_collection_Scientific_Abstracts             = df[df.collection== 'Scientific_Abstracts']
df_collection_Consolidated_Translations        = df[df.collection== 'Consolidated_Translations']
df_collection_Misc                             = df[df.collection== 'Misc']

#df_majority = df[df.balance==0]
#df_minority = df[df.balance==1]
# Ref: https://elitedatascience.com/imbalanced-classes
 
    
df_collection_General_CIA_Records = resample(df_collection_General_CIA_Records, 
                                             replace=True,    # sample with replacement
                                             n_samples=15000,     # to match number of values in each class
                                             random_state=123) # reproducible results

df_collection_NGA_Records   = resample(df_collection_NGA_Records, 
                                             replace=True,    # sample with replacement
                                             n_samples=15000,     # to match number of values in each class
                                             random_state=123) # reproducible results

df_collection_Scientific_Abstracts = resample(df_collection_Scientific_Abstracts, 
                                             replace=True,    # sample with replacement
                                             n_samples=15000,     # to match number of values in each class
                                             random_state=123) # reproducible results

df_collection_Consolidated_Translations = resample(df_collection_Consolidated_Translations, 
                                             replace=True,    # sample with replacement
                                             n_samples=15000,     # to match number of values in each class
                                             random_state=123) # reproducible results

df_collection_Misc  = resample(df_collection_Misc, 
                                replace=True,    # sample with replacement
                                n_samples=15000,     # to match number of values in each class
                                random_state=123) # reproducible results

    
# Combine all the class with equal number of values
df_sampled = pd.concat([df_collection_General_CIA_Records, df_collection_NGA_Records, df_collection_Scientific_Abstracts,
                       df_collection_Consolidated_Translations, df_collection_Misc])
 

# check for the class balance
df_sampled['collection'].value_counts()
```

```python
# change df with sampled one
df= df_sampled
```
